```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
def split_normalization_and_standardization(path_of_csv: str, random_state: int, test_size: float):
    df = pd.read_csv(path_of_csv)
    
    X = df.iloc[:, :-2]
    y = np.array(df['label'])
    source = np.array(df["source"])
    
    # Split into train & test
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = test_size, random_state=random_state)
    # Split into train & validation
    X_train, X_validation, y_train, y_validation = train_test_split(X_train, y_train, test_size = test_size, random_state=random_state)
    
    # Normalization
    min_max_scaler = MinMaxScaler()
    X_train_norm = pd.DataFrame(min_max_scaler.fit_transform(X_train))
    X_validation_norm = pd.DataFrame(min_max_scaler.fit_transform(X_validation))
    X_test_norm = pd.DataFrame(min_max_scaler.fit_transform(X_test))
    
    # Standardization
    standard_scaler = StandardScaler()
    X_train_scaled = pd.DataFrame(standard_scaler.fit_transform(X_train_norm))
    X_validation_scaled = pd.DataFrame(standard_scaler.fit_transform(X_validation_norm))
    X_test_scaled = pd.DataFrame(standard_scaler.fit_transform(X_test_norm))
    
    return X_train_scaled, X_validation_scaled, X_test_scaled, source
```

```python
X_train, X_validation, X_test, source = split_normalization_and_standardization("./data/dataset.csv", 66, 0.2)
```
